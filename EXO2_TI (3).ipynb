Avec l'image du Loup-noir

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab import drive

drive.mount('/content/drive')

# Charger l'image
img_path = '/content/drive/MyDrive/TP1_Hist-OptPxl/Loup-noir.jpg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# Afficher l'image originale et son histogramme
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title('Image originale')
plt.subplots_adjust(hspace=0.5)

plt.subplot(2,2,2)
plt.hist(img.ravel(),256,[0,256])
plt.title('Histogramme original')
# fonction histogramme sans hist
# Calculer l'histogramme
#hist = [0] * 256
#for i in range(img.shape[0]):
 #   for j in range(img.shape[1]):
  #      hist[img[i,j]] += 1
  

# Afficher l'histogramme
#plt.subplot(2,2,3)
#plt.bar(range(256), hist)
#plt.title('Histogramme')

# Trouver les valeurs minimale et maximale de l'histogramme
min_val = np.min(img)
max_val = np.max(img)

# Étirer l'histogramme
a = 0
b = 255
img_stretched = (img - min_val) * ((b - a) / (max_val - min_val)) + a
img_stretched = np.clip(img_stretched, a, b).astype(np.uint8)

# Afficher l'image étirée et son histogramme
plt.subplot(2,2,3)

plt.hist(img_stretched.ravel(),256,[0,256])
plt.title('Histogramme étiré')

plt.subplot(2,2,4)
plt.imshow(img_stretched, cmap='gray')
plt.title('Image étirée')

# Mesurer le contraste avec la mesure de Michelson
C_M = (np.max(img_stretched) - np.min(img_stretched)) / (np.max(img_stretched) + np.min(img_stretched))
print(f"Mesure de Michelson : {C_M}")

# Mesurer le contraste avec la mesure RMS
C_RMS = np.sqrt(np.mean(np.square(img_stretched))) / np.mean(img_stretched)
print(f"Mesure RMS : {C_RMS}")

# Chemin complet du fichier à enregistrer dans Google Drive
file_path = '/content/drive/MyDrive/TP1_Hist-OptPxl/LoupEtirer.jpg'

# Enregistrer l'image étirée dans Google Drive
cv2.imwrite(file_path, img_stretched)
plt.show()
```
