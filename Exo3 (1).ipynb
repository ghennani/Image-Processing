Avec l'image aqui.com

```python
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
drive.mount('/content/drive')
#Python Imaging Library
import cv2
from google.colab import drive
drive.mount('/content/drive')
# Charger l'image
path = '/content/drive/MyDrive/TP1_Hist-OptPxl/aqui.pgm'
img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img)
plt.title("image originale")
plt.show()
#calcul de l'histogramme
hist, bins = np.histogram(img,256,[0,256])
# Cration du diagramme en barres
plt.bar(bins[:-1], hist, width = 1)
# Ajout d'un titre et d'tiquettes pour les axes
plt.title("Histogramme de l'image originale")
plt.xlabel("Niveaux de gris")
plt.ylabel("Nombre de pixels")
plt.show()
#calcule de l'histogramme cumulative
hist_norm = cv2.calcHist([img], [0], None, [256], [0, 256]) / (img.shape[0] * img.shape[1])
cumulative_hist_norm = np.cumsum(hist_norm)

#Crer la transformation
transformation = np.round(cumulative_hist_norm * 255).astype(np.uint8)

#Appliquer la transformation  l'image
img_equalized = np.zeros_like(img)
for i in range(img.shape[0]):
  for j in range(img.shape[1]):
    img_equalized[i, j] = transformation[img[i, j]]
# on a put utiliser cette ligne pour l'egalisation img_equalized = cv2.equalizeHist(image)
#Afficher l'image galise et son histogramme
plt.subplot(2,2,3)
plt.imshow(img_equalized, cmap='gray')
plt.title('Image galise')

plt.subplot(2,2,4)
plt.hist(img_equalized.ravel(),256,[0,256])
plt.title('Histogramme galis')
# Chemin complet du fichier  enregistrer dans Google Drive
file_path = '/content/drive/MyDrive/TP1_Hist-OptPxl/aquiEgaliser.pgm'

# Enregistrer l'image tire dans Google Drive
cv2.imwrite(file_path, img_equalized)
plt.show()

```
